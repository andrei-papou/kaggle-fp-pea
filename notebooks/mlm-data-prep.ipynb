```python
def _generate_discourse_mlm_df(df: pd.DataFrame, exclude_id_set: t.Set[str], out_path: Path):
    mlm_df = df[~df['id'].isin(exclude_id_set)].sample(n=20000)
    with open(out_path, 'w') as f:
        for _, row in tqdm(mlm_df.iterrows(), total=len(mlm_df)):
            f.write(row['discourse_text'].strip() + '\n')

_generate_discourse_mlm_df(fp2021_df, exclude_id_set=used_essay_id_set, out_path=FP_0P925_DIR / 'mlm_discourse.txt')
```
